# Open and Close Price Difference 

```python
df['diff'] = df['Open'] - df['Close']
df_diff = df.resample('W').mean()
df_diff.tail(10)

df_18_diff = df[pd.Timestamp('2018-01-01'):pd.Timestamp('2018-12-31')]
df_19_diff = df[pd.Timestamp('2019-01-01'):pd.Timestamp('2019-12-31')]
df_20_diff = df[pd.Timestamp('2020-01-01'):pd.Timestamp('2020-12-31')]
df_21_diff = df[pd.Timestamp('2021-01-01'):pd.Timestamp('2021-12-31')]
df_22_diff = df[pd.Timestamp('2022-01-01'):pd.Timestamp('2022-12-31')]
```

```python
f, ax = plt.subplots(2, 2, figsize=(10,10), sharex=True, sharey=True)

ax[0,0].scatter(df_diff.loc['2018-01-01':'2018-12-31'].index, df_diff.loc['2018-01-01':'2018-12-31']['diff']
, color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('2018');

ax[0,1].scatter(df_diff.loc['2019-01-01':'2019-12-31'].index, df_diff.loc['2019-01-01':'2019-12-31']['diff']
, color='r')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('2019');

ax[1,0].scatter(df_diff.loc['2020-01-01':'2020-12-31'].index, df_diff.loc['2020-01-01':'2020-12-31']['diff']
, color='r')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('2020');

ax[1,1].scatter(df_diff.loc['2021-01-01':'2021-12-31'].index, df_diff.loc['2021-01-01':'2021-12-31']['diff']
, color='r')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('2021');




```

# Daily Returns (Moving Windows)

```python
daily_close_df = df[['Adj Close']]

# Daily returns
daily_pct_change_df = daily_close_df.pct_change()

daily_pct_change_df.fillna(0, inplace=True)

daily_pct_change_df.head()
```

```python
daily_pct_change_df.hist(bins=50)
```

```python
import seaborn as sns
sns.set()
```

```python
f, axes = plt.subplots(figsize=(12, 7))

sns.distplot(daily_pct_change_df['Adj Close'], color="b");
```

# Volatility 

import numpy as np

min_periods = 75 

vol = daily_pct_change_df.rolling(min_periods).std() * np.sqrt(min_periods) 

vol.fillna(0,inplace=True)

vol.tail()

```python
vol.plot(figsize=(10, 8))
plt.show()
```

# Rolling Means

```python
df_adj_close_px = df['Adj Close']
# Short moving window rolling mean
df['42'] = df_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
df['252'] = df_adj_close_px.rolling(window=252).mean()

df[['Adj Close', '42', '252']].plot(title="Rolling Means")
plt.show()
```
